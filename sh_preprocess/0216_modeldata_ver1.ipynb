중분류, 소분류, 세분류별로 score 만들기! -> 였는데 중분류만 잘라서 드릴게요^.^

```python
#중분류
totallist = []
for pdlist in tqdm(pgm.sorted_상품중분류명):
    if type(pdlist) == list:
        totallist += pdlist
        
totaldict = {}
for prd in tqdm(totallist):
    if prd not in totaldict:
        totaldict[prd] = 1
    else:
        totaldict[prd] += 1
```

```python
outlist = []
for pdlist in outlier.sorted_상품중분류명:
    if type(pdlist) == list:
        outlist += pdlist

outdict = {}
for prd in outlist:
    if prd not in outdict:
        outdict[prd] = 1
    else:
        outdict[prd] += 1
```

```python
ratio = {}
for prd in totaldict:
    if prd in outdict:
        ratio[prd] = outdict[prd] / totaldict[prd]
    else:
        ratio[prd] = 0
```

```python
joongratio = ratio
joong = pd.DataFrame.from_dict(ratio, orient='index').reset_index()
```

```python
joong = joong.rename(columns={'index':'상품중분류명', 0:'lim_ratio1'})
# joong.to_pickle('0212_1526_joong_lim_ratio.pkl')
```

```python
# 상품명을 통해서 브랜드명, 가격을 함께 뽑아보아요ㅎㅎ
# 상품명을 key로, 중분류score를 value로 갖는 딕셔너리를 컬럼으로 넣어줍니다.
def makejoong(row):
    valueslist = row['comb_상품중분류명']
    prdslist = row['comb_판매상품명']
    scoredict = {}
    for i, value in enumerate(valueslist):
        scoredict[prdslist[i]] = joongratio[value]
    return scoredict
```

```python
pgm['joongscore'] = pgm.apply(lambda row : makejoong(row), axis=1)
```

```python
from collections import Counter

# score가 가장 높은 상품의 브랜드를 찾아주는 함수!
def makebrandcol(row): # 가장 큰 값으로 채우기
    brands = row['comb_브랜드명']
    prds = row['comb_판매상품명']
    scoredict = row['joongscore']
    collist = [] # 컬럼의 내용이 되는 리스트
    if len(scoredict) >= 3: # 상품 개수가 3개 이상
        for prdset in Counter(scoredict).most_common(3): # 각각의 상품에 대해서
            prdnm = prdset[0] # prdnm = 상품명
            index = prds.index(prdnm) # 상품명의 순서 추출
            collist.append(brands[index]) # 해당 순서의 브랜드 리스트에 넣어줍니다
    else: # 상품 개수가 3개 미만
        for prdset in Counter(scoredict).most_common(1):
            prdnm = prdset[0]
            index = prds.index(prdnm)
            collist.append(brands[index])
            collist.append(brands[index])
            collist.append(brands[index])
    return collist



# score가 가장 높은 상품의 가격을 찾아주는 함수!
def makepricecol(row): # 가장 큰 값으로 채우기
    prices = row['상품판매가']
    prds = row['comb_판매상품명']
    scoredict = row['joongscore']
    collist = [] # 컬럼의 내용이 되는 리스트
    if len(scoredict) >= 3: # 상품 개수가 3개 이상
        for prdset in Counter(scoredict).most_common(3):
            prdnm = prdset[0]
            index = prds.index(prdnm)
            collist.append(prices[index])
    else: # 상품 개수가 3개 미만
        for prdset in Counter(scoredict).most_common(1):
            prdnm = prdset[0]
            index = prds.index(prdnm)
            collist.append(prices[index])
            collist.append(prices[index])
            collist.append(prices[index])
    return collist


# score가 가장 높은 상품의 score만 남기는 함수!
def makescorecol(row): # 가장 큰 값으로 채우기
    scoredict = row['joongscore']
    collist = [] # 컬럼의 내용이 되는 리스트
    if len(scoredict) >= 3: # 상품 개수가 3개 이상
        for prdset in Counter(scoredict).most_common(3): # 각각의 상품에 대해서
            prdnm = prdset[0] # prdnm = 상품명
            collist.append(scoredict[prdnm])
    else: # 상품 개수가 3개 미만
        for prdset in Counter(scoredict).most_common(1):
            prdnm = prdset[0]
            collist.append(scoredict[prdnm])
            collist.append(scoredict[prdnm])
            collist.append(scoredict[prdnm])
    return collist
```

```python
pgm['brands'] = pgm.apply(lambda row : makebrandcol(row), axis=1)
pgm['prices'] = pgm.apply(lambda row : makepricecol(row), axis=1)
pgm['scores'] = pgm.apply(lambda row : makescorecol(row), axis=1)
```
